### Analysis of Sexual Violence Types with Data Filtering Functions 

```python
class Type_of_SV:
    def __init__(self, data):
        if 'Type of SV' not in data.columns:
            raise ValueError("DataFrame must contain a 'Type of SV' column.")
        self.data = data

    def filter_by_Rape(self):
        return self.data[self.data['Type of SV'] == 'Rape']

    def filter_by_SexualAssault(self):
        return self.data[self.data['Type of SV'] == 'SexualAssault']

    def filter_by_AttemptedSexualAssault(self):
        return self.data[self.data['Type of SV'] == 'AttemptedSexualAssault']

    def filter_by_AttemptedSexualAssault_and_Rape(self):
        return self.data[self.data['Type of SV'] == 'AttemptedSexualAssault, Rape']

    def filter_by_CuttingOff_InjuringSexualOrgans(self):
        return self.data[self.data['Type of SV'] == 'CuttingOff/InjuringSexualOrgans']

    def filter_by_Forced_ChildMarriage(self):
        return self.data[self.data['Type of SV'] == 'Forced/ChildMarriage']

    def filter_by_ForcedToWitnessSexualViolence_Rape(self):
        return self.data[self.data['Type of SV'] == 'ForcedToWitnessSexualViolence, Rape']

    def filter_by_NotAnswered(self):
        return self.data[self.data['Type of SV'] == 'NotAnswered']

    def filter_by_Other(self):
        return self.data[self.data['Type of SV'] == 'Other']

    def filter_by_Other_Rape(self):
        return self.data[self.data['Type of SV'] == 'Other, Rape']

    def filter_by_Rape_SexualAssault(self):
        return self.data[self.data['Type of SV'] == 'Rape, SexualAssault']

    def filter_by_Rape_SexualHarassment(self):
        return self.data[self.data['Type of SV'] == 'Rape, SexualHarassment']

    def filter_by_Rape_TransactionalSex(self):
        return self.data[self.data['Type of SV'] == 'Rape, TransactionalSex']

    def filter_by_Rape_UnspecifiedActSV(self):
        return self.data[self.data['Type of SV'] == 'Rape, UnspecifiedActSV']

    def filter_by_SexualHarassment(self):
        return self.data[self.data['Type of SV'] == 'SexualHarassment']

    def filter_by_StrippedNaked_TornClothes(self):
        return self.data[self.data['Type of SV'] == 'StrippedNaked/TornClothes']

    def filter_by_ThreatOfSexualViolence(self):
        return self.data[self.data['Type of SV'] == 'ThreatOfSexualViolence']

    def filter_by_TransactionalSex(self):
        return self.data[self.data['Type of SV'] == 'TransactionalSex']

    def filter_by_UnspecifiedActSV(self):
        return self.data[self.data['Type of SV'] == 'UnspecifiedActSV']

    def filter_by_UnspecifiedActSV_UnwantedSexualTouching(self):
        return self.data[self.data['Type of SV'] == 'UnspecifiedActSV, UnwantedSexualTouching']

    def filter_by_UnwantedSexualTouching(self):
        return self.data[self.data['Type of SV'] == 'UnwantedSexualTouching']
```

```python
sv_filter = Type_of_SV(data)

rape_data = sv_filter.filter_by_Rape()
sexual_assault_data = sv_filter.filter_by_SexualAssault()
attempted_sexual_assualt_data = sv_filter.filter_by_AttemptedSexualAssault()
attempted_sexual_assualt_and_rape_data = sv_filter.filter_by_AttemptedSexualAssault_and_Rape()
cuttingoff_injuring_sexualorgans_data = sv_filter.filter_by_CuttingOff_InjuringSexualOrgans()
forced_childmarriage_data = sv_filter.filter_by_Forced_ChildMarriage()
forced_to_witness_sexualviolence_rape_data = sv_filter.filter_by_ForcedToWitnessSexualViolence_Rape()
not_answered_data = sv_filter.filter_by_NotAnswered()
other_data = sv_filter.filter_by_Other()
other_rape_data = sv_filter.filter_by_Other_Rape()
rape_sexual_assualt_data = sv_filter.filter_by_Rape_SexualAssault()
rape_sexual_harassment_data = sv_filter.filter_by_Rape_SexualHarassment()
rape_transactionalsex_data = sv_filter.filter_by_Rape_TransactionalSex()
rape_unspecified_actsv_data = sv_filter.filter_by_Rape_UnspecifiedActSV()
sexual_harassment_data = sv_filter.filter_by_SexualHarassment()
strippednaked_tornclothes_data = sv_filter.filter_by_StrippedNaked_TornClothes()
threat_of_sexualviolence = sv_filter.filter_by_ThreatOfSexualViolence()
transactional_sex_data = sv_filter.filter_by_TransactionalSex()
unspecifiedact_sv_data = sv_filter.filter_by_UnspecifiedActSV()
UnspecifiedActSV_UnwantedSexualTouching_data = sv_filter.filter_by_UnspecifiedActSV_UnwantedSexualTouching()
UnwantedSexualTouching_data = sv_filter.filter_by_UnwantedSexualTouching()
```

### Plotting
#### Count the Types of Sexual Violences rate

```python
class Visualize:

    @staticmethod
    def filter_by_type(data, attribute_name):
        if 'Type of SV' not in data.columns:
            raise ValueError("DataFrame must contain a 'Type of SV' column.")
        return data[data['Type of SV'] == attribute_name]

# Assuming 'data' is a DataFrame with the 'Type of SV' column
attribute_names = ['AttemptedSexualAssault','AttemptedSexualAssault, Rape','CuttingOff/InjuringSexualOrgans',
                   'Forced/ChildMarriage','ForcedToWitnessSexualViolence, Rape','NotAnswered','Other','Other, Rape','Rape',
                   'Rape, SexualAssault','Rape, SexualHarassment','Rape, TransactionalSex','Rape, UnspecifiedActSV',
                   'SexualAssault','SexualHarassment','StrippedNaked/TornClothes','ThreatOfSexualViolence','TransactionalSex',
                   'UnspecifiedActSV','UnspecifiedActSV, UnwantedSexualTouching','UnwantedSexualTouching']

counts = {}
for attr in attribute_names:
    filtered_data = Visualize.filter_by_type(data, attr)
    counts[attr] = len(filtered_data)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(list(counts.values()), list(counts.keys()), marker='d', linestyle='-', color='r')
plt.xlabel('Count')
plt.ylabel('Type of SV')
plt.title('Counts of Each Type of SV')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.bar(counts.keys(), counts.values(),color='r')
plt.xlabel('Type of SV')
plt.ylabel('Count')
plt.title('Counts of Each Type of SV')
plt.xticks(rotation=90)
plt.show()
```

### Data Aggregation
##### Group the data by 'Country ' and 'Type of SV' and count occurrences.

```python
sv_counts = data.groupby(['Country', 'Type of SV']).size().unstack(fill_value=0)
```

### A heatmap will give you a visual representation of counts where the intensity of color indicates the number of occurrences.

```python
plt.figure(figsize=(14, 10))
sns.heatmap(sv_counts, cmap="YlGnBu", annot=True, fmt="d", cbar=True)
plt.title('Types of Sexual Violence by Country')
plt.xlabel('Type of Sexual Violence')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Plotting the top 5 SV types per country

```python
top_sv_per_country = sv_counts.apply(lambda x: x.nlargest(5).index.tolist(), axis=1)

for country, top_sv in top_sv_per_country.items():
    plt.figure(figsize=(15, 8))
    
    # Extract the counts for the top SV types
    sv_data = sv_counts.loc[country, top_sv]
    
    # Plot as horizontal bar plot
    sv_data.plot(kind='barh', color='red')
    
    plt.title(f'Top 5 Types of Sexual Violence in {country}')
    plt.xlabel('Count')
    plt.ylabel('Type of Sexual Violence')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```
